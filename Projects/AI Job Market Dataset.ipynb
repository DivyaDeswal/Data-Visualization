## find relation between job titles and company size

```python
data=df.groupby('Job_Title')['Company_Size'].value_counts().unstack()
print(data)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
plt.figure(figsize=(12,10))
data_melted = data.reset_index().melt(id_vars='Job_Title', var_name='Company_Size', value_name='Count')
sns.barplot(x='Job_Title', y='Count', hue='Company_Size', data=data_melted, palette=colors)
plt.title('Distribution of Job Title by Company Size')
plt.xlabel('Job Title')
plt.ylabel('Count')
plt.xticks(rotation=45)
plt.legend(title='Company Size', bbox_to_anchor=(0, 1))
plt.show()
```

```python
plt.figure(figsize=(12,10))
data_melted = data.reset_index().melt(id_vars='Job_Title', var_name='Company_Size', value_name='Count')
sns.barplot(x='Job_Title', y='Count', hue='Company_Size', data=data_melted, palette=colors)
plt.title('Distribution of Job Title by Company Size')
plt.xlabel('Job Title')
plt.ylabel('Count')
plt.xticks(rotation=45)
plt.legend(title='Company Size', bbox_to_anchor=(0, 1))
for index, value in enumerate(data_melted):
    plt.text(index,value + 2, str(value), color='black', ha="right", size='small')
plt.show()
```

## plot AI risk with respect to company size

```python
df['AI_Adoption_Level'].value_counts()
```

```python
x=df.groupby(['Company_Size','AI_Adoption_Level']).size().unstack()
x
```

```python
plt.figure(figsize=(10,8))
data_melt=x.reset_index().melt(id_vars='Company_Size',var_name='AI_Adoption_Level',value_name='Count')
sns.barplot(data=data_melt,x='AI_Adoption_Level',y='Count',hue='Company_Size')
plt.title('AI adoption with different company size')
plt.xlabel('AI adoption level')
plt.ylabel('Counts')
plt.xticks(rotation=35)
for p in plt.gca().patches:
    plt.text(p.get_x() + p.get_width() / 2, p.get_height() + 2, 
             '{:1.0f}'.format(p.get_height()), ha='center', va='center',color='red')
plt.show()
```

```python
plt.figure(figsize=(10,8))
ax=x.plot(kind='bar', stacked=True, colormap='tab20')
plt.title('AI adoption with different company size')
plt.xlabel('AI adoption level')
plt.ylabel('Counts')
plt.xticks(rotation=35)
for container in ax.containers:
    ax.bar_label(container, label_type='center', fmt='%d')
plt.show()
```
